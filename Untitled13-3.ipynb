```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

%matplotlib inline

class ViratSmartCityAnalyzer:
    def __init__(self, fps=30):
        self.fps = fps
        self.data = None
        self.evt_df = None
        self.type_map = {1: 'Pedestrian', 2: 'Car', 4: 'Bicycle', 0: 'Other'}
        self.event_map = {
            1: 'Loading', 2: 'Unloading', 3: 'Opening Trunk', 4: 'Closing Trunk',
            5: 'Getting In', 6: 'Getting Out', 7: 'Entering', 8: 'Exiting',
            9: 'Walking', 10: 'Turning', 11: 'U-Turn', 12: 'Starting', 13: 'Stopping', 14: 'Running'
        }

    def load_data(self, obj_path, evt_path):
        if os.path.exists(obj_path):
            df = pd.read_csv(obj_path, sep=' ', header=None)
            self.data = df[[0, 2, 3, 4, 5, 6, 7]].rename(columns={
                0: 'obj_id', 2: 'frame_id', 3: 'x', 4: 'y', 5: 'w', 6: 'h', 7: 'type_id'
            })
            self.data['type'] = self.data['type_id'].map(self.type_map).fillna('Other')
            self.data['sec'] = (self.data['frame_id'] / self.fps).astype(int)

        if os.path.exists(evt_path):
            edf = pd.read_csv(evt_path, sep=' ', header=None)
            self.evt_df = edf[[0, 1, 3, 5]].rename(columns={
                0: 'evt_id', 1: 'evt_type_id', 3: 'start_frame', 5: 'frame_id'
            })
            self.evt_df['event'] = self.evt_df['evt_type_id'].map(self.event_map)
            self.evt_df['sec'] = (self.evt_df['frame_id'] / self.fps).astype(int)

    def load_yolo_data(self, yolo_folder, img_w=1280, img_h=720):
        """兼容 YOLO 数据的转换器"""
        import glob
        from pathlib import Path
        records = []
        files = glob.glob(os.path.join(yolo_folder, "*.txt"))
        
        for f in files:
            try:
                frame_id = int(Path(f).stem) # 假设 YOLO 结果的文件名是帧号
                with open(f, 'r') as file:
                    for line in file:
                        p = line.split()
                        if len(p) >= 5:
                            # YOLO 格式: class x_center y_center w h
                            records.append({
                                'frame_id': frame_id,
                                'sec': int(frame_id / self.fps),
                                'type': self.type_map.get(int(p[0]), 'Other'),
                                'x': float(p[1]) * img_w,
                                'y': float(p[2]) * img_h,
                                'obj_id': f"{frame_id}_{p[1]}" # YOLO 无追踪ID，生成临时ID用于绘图
                            })
            except: continue
                
        self.data = pd.DataFrame(records)
        self.evt_df = None # 告诉绘图系统，YOLO 没有高阶事件数据
        print(f"YOLO data loaded successfully! A total of {len(self.data)} target coordinates have been extracted.")

    def plot_analysis(self):
        if self.data is None:
            return
        
        fig = plt.figure(figsize=(20, 15))
        
        plt.subplot(2, 2, 1)
        unique_objs = self.data.drop_duplicates('obj_id')
        sns.countplot(data=unique_objs, x='type', hue='type', palette='viridis', legend=False)
        plt.title("01 Function: Unique Object Distribution")
        
        plt.subplot(2, 2, 2)
        flow = self.data.groupby('sec').size()
        plt.plot(flow.index, flow.values, color='tab:red', lw=2)
        plt.fill_between(flow.index, flow.values, alpha=0.2, color='tab:red')
        plt.title("04 Function: Real-time Traffic Intensity")
        plt.xlabel("Seconds")
        
        plt.subplot(2, 2, 3)
        plt.hexbin(self.data['x'], self.data['y'], gridsize=40, cmap='YlOrRd', mincnt=1)
        plt.colorbar(label='Activity Density')
        plt.gca().invert_yaxis()
        plt.title("04 Function: REAL Spatial Activity Heatmap")
        
        if self.evt_df is not None:
            plt.subplot(2, 2, 4)
            evt_counts = self.evt_df.drop_duplicates('evt_id')['event'].value_counts()
            evt_counts.plot(kind='barh', color='teal')
            plt.title("Enhanced Function: Event/Activity Analysis")
            
        plt.tight_layout()
        plt.show()

analyzer = ViratSmartCityAnalyzer()
obj_file = "VIRAT_S_000001.viratdata.objects.txt"
evt_file = "VIRAT_S_000001.viratdata.events.txt"

analyzer.load_data(obj_file, evt_file)
analyzer.plot_analysis()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# 确保在 Jupyter 中直接显示图表
%matplotlib inline

class ViratInteractionAnalyzer:
    def __init__(self):
        self.obj_df = None
        self.map_df = None
        # VIRAT 官方事件代码映射
        self.event_map = {
            1: 'Loading', 2: 'Unloading', 3: 'Opening Trunk', 
            4: 'Closing Trunk', 5: 'Getting In', 6: 'Getting Out', 
            7: 'Entering Facility', 8: 'Exiting Facility',
            9: 'Walking', 10: 'Turning', 11: 'U-Turn', 
            12: 'Starting', 13: 'Stopping', 14: 'Running'
        }

    def load_and_map_events(self, obj_path, map_path):
        if not (os.path.exists(obj_path) and os.path.exists(map_path)):
            print("fail")
            return

        # 1. 提取所有物体的坐标 (用于精确定位)
        self.obj_df = pd.read_csv(obj_path, sep=' ', header=None)
        self.obj_df = self.obj_df[[0, 2, 3, 4, 5, 6]].rename(columns={
            0: 'obj_id', 2: 'frame', 3: 'x', 4: 'y', 5: 'w', 6: 'h'
        })
        # 计算物体的中心点坐标
        self.obj_df['cx'] = self.obj_df['x'] + self.obj_df['w'] / 2
        self.obj_df['cy'] = self.obj_df['y'] + self.obj_df['h'] / 2

        # 2. 解析 mapping.txt (事件-物体关联指示矩阵)
        # 使用 \s+ 处理文件中不规则的多个空格
        self.map_df = pd.read_csv(map_path, sep='\s+', header=None)
        
        interactions = []
        for index, row in self.map_df.iterrows():
            evt_id = row[0]
            evt_type = row[1]
            start_frame = row[3]
            end_frame = row[4]
            
            # 从第 6 列开始，每一列代表一个物体是否参与了该事件 (1代表参与，0代表未参与)
            for i, col_idx in enumerate(range(6, len(row))):
                if row[col_idx] == 1:
                    interactions.append({
                        'evt_id': evt_id,
                        'evt_name': self.event_map.get(evt_type, 'Other'),
                        'obj_id': i,  # i 就是物体的真实 ID
                        'start_frame': start_frame,
                        'end_frame': end_frame
                    })
        
        self.interaction_df = pd.DataFrame(interactions)
        print(f"Found {len(self.interaction_df)} human/vehicle spatial interaction records.")

    def plot_interaction_hotspots(self):
        """绘制交互行为空间发生热图"""
        if self.obj_df is None or self.interaction_df.empty:
            return
            
        # 将事件与当时的物体坐标合并
        merged = pd.merge(self.interaction_df, self.obj_df, on='obj_id')
        
        # 严谨过滤：只保留事件真正发生期间的坐标
        merged = merged[(merged['frame'] >= merged['start_frame']) & (merged['frame'] <= merged['end_frame'])]
        
        # 计算每次事件发生的中心位置（取整个行为发生期间的中位数坐标）
        evt_locations = merged.groupby(['evt_id', 'evt_name'])[['cx', 'cy']].median().reset_index()
        
        # --- 开始绘图 ---
        plt.figure(figsize=(14, 8))
        
        # 使用气泡散点图展示各种事件的发生位置
        sns.scatterplot(data=evt_locations, x='cx', y='cy', hue='evt_name', 
                        palette='tab10', s=250, edgecolor='white', linewidth=2, alpha=0.9)
        
        plt.gca().invert_yaxis()  # 匹配真实视频图像的坐标系 (Y轴向下)
        plt.title("Advanced Function: Human-Vehicle Interaction Map", fontsize=16, fontweight='bold')
        plt.xlabel("X Coordinate (Pixels)", fontsize=12)
        plt.ylabel("Y Coordinate (Pixels)", fontsize=12)
        
        # 将图例移到图表外，避免遮挡数据点
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title="Event Types", fontsize=11, title_fontsize=13)
        plt.grid(True, linestyle='--', alpha=0.4)
        plt.tight_layout()
        plt.show()

# --- 直接运行逻辑 ---
interaction_analyzer = ViratInteractionAnalyzer()

# 确保填入你正确的文件名
obj_file = "VIRAT_S_000001.viratdata.objects.txt"
map_file = "VIRAT_S_000001.viratdata.mapping.txt"

interaction_analyzer.load_and_map_events(obj_file, map_file)
interaction_analyzer.plot_interaction_hotspots()
```
